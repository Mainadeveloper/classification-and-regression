```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
```

```python
df=pd.read_csv('C:/classisificatn and regressin/House_Price.csv',header=0)
```

```python
df.head()
```

```python
Xvar=sm.add_constant(df['room_num'])
# Adding a constant term (intercept) to the independent variable 'room_num'
# This prepares the data for the OLS regression model by including an intercept term.
```

```python
# Fitting the Ordinary Least Squares (OLS) regression model
lmvar=sm.OLS(df['price'],Xvar).fit()
# Here, 'price' is the dependent variable, and 'Xvar' (which includes the intercept and 'room_num') is the independent variable.
```

```python
lmvar.summary()
# This provides a detailed overview of the regression results, including coefficients, R-squared, and p-values.
```

#Overall, the regression results indicate that room_num has a statistically significant effect on price, explaining nearly half of the variance in property prices. However, there are likely other factors influencing prices that are not captured by this model, as indicated by the R-squared value.
